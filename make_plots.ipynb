```python
def draw_percent_diff(dataframe, col_to_plot, comparison=None, ax=None, vscale=(None,None)):

  title = col_to_plot


  if 'eig_spec' in dataframe.columns:
    df3 = pd.DataFrame(dataframe['eig_spec'].to_list(), columns=['eig{:d}'.format(i) for i in range(36)])
    dataframe = dataframe.join(df3)

  if comparison is not None:
    dataframe['abs_error_gs_comp'] = comparison.to_numpy()
    dataframe['err_reduction'] = dataframe.apply(lambda row: ((row.abs_error_gs_comp - row.abs_error_gs)/((row.abs_error_gs_comp+row.abs_error_gs)/2))*100, axis=1)
    col_to_plot = 'err_reduction'
    title = 'Percent difference from standard reference'
  
  heat_data = dataframe.pivot('g', 'pb', col_to_plot)
#   mask_idx = dataframe.loc[(dataframe[col_to_plot] < 0.1) & (dataframe[col_to_plot] > -0.1) ]
#   mask = dataframe.copy()
#   for val in mask[col_to_plot]:
#     if (val < 3.1) | (val > 2.9):
#         val = 1
#     else:
#         val = 0
  # fig = plt.figure(figsize=(8,8))
  h = sns.heatmap(heat_data, 
                  xticklabels=['{:.2f}'.format(x) for x in g_vals], 
                  yticklabels=['{:.2f}'.format(y) for y in pb_vals],
                  # annot=np.reshape(gs_mesh_data.index, [30,30]), fmt='d', annot_kws={"size": 12}, 
                  robust=True,
                  # vmin=-150,vmax=150)
                  vmin=vscale[0],vmax=vscale[1],
                  ax=ax)
#   m = sns.heatmap(heat_data, mask=mask.pivot('g', 'pb', col_to_plot),ax=ax)
  ax.set_title(title)
```

```python
def draw_plot_evec_comps(dataframe, index):
  eig_v = dataframe['ci_eig_vs'].loc[index]
  g = dataframe['g'].loc[index]
  pb = dataframe['pb'].loc[index]

  fig = plt.figure(figsize=(16,4))
  sns.barplot(x=['v{:d}'.format(i) for i in range(36)], y=eig_v[:,0]*eig_v[:,0])
  plt.text(0.68, 0.6, 
         '(g, pb) = ({:0.4f}, {:0.4f})'.format(g, pb),
         color='red',
         fontsize=14, 
         transform=fig.axes[0].transAxes)
  # plt.savefig('eig_comps_{:d}.png'.format(index))
```

```python
def generate_data(g_vals, pb_vals):
  df_list = []
  #states = pyci.init_states(4,4)
  for g in g_vals:
    hme = pyci.matrix(4,4, 0.0, 1.0, g, 0.0) 
    w1, v1 = np.linalg.eig(hme)
    w1_sort = np.sort(w1)
    sort_idx = np.argsort(w1)
    v1 = v1[:,sort_idx]
    w2, v2 = np.linalg.eigh(hme)
    #print(w1_sort-w2)
    w, v = np.linalg.eigh(hme)
    w=w1_sort
    v=v1
    df = pd.DataFrame({'g':g, 'pb':0.0, 'hme':[hme], 'eigw':[w], 'ci_eig_vs':[v]})
    df_list.append(df)
  print('done g')
  for pb in pb_vals:
    hme = pyci.matrix(4,4, 0.0, 1.0, 0.0, pb) 
    w1, v1 = np.linalg.eig(hme)
    w1_sort = np.sort(w1)
    sort_idx = np.argsort(w1)
    v1 = v1[:,sort_idx]
    w2, v2 = np.linalg.eigh(hme)
    #print(w1_sort-w2)
    w, v = np.linalg.eigh(hme)
    w=w1_sort
    v=v1
    df = pd.DataFrame({'g':0.0, 'pb':pb, 'hme':[hme], 'eigw':[w], 'ci_eig_vs':[v]})
    df_list.append(df)   
  print('done pb') 
  print('df_list')
  data_df = pd.concat(df_list, ignore_index=True)

  return data_df
```

```python
sns.set_context("paper", 
                rc={"font.size":20,"axes.titlesize":15,"axes.labelsize":15, "lines.linewidth":2.5},
                font_scale=1.5)
def plot_expects(data):
    fig = plt.figure(figsize=(8,4))
    plt.axhline(0, linestyle='--', color='red')
    sns.lineplot(x='s', y=data['E_gs'][0]-data['E_gs'], data=data)
    sns.lineplot(x='s', y=data['s_expect'][0]-data['s_expect'], data=data)
    plt.legend(['diff = 0','E_gs(s=0)-E_gs(s)', 'SS(s=0)-SS(s)'])
    plt.ylabel('Divergence from s=0')
plot_expects(ss_flow_1)
plt.title('Single standard reference\nd=1.0, g=0.5, pb=0.0\nnholes=4, nparticles=4')

plot_expects(ss_flow_2)
plt.title('Single standard reference\nd=1.0, g=0.5, pb=0.1\nnholes=4, nparticles=4')

plot_expects(ss_flow_3)
plt.title('Single standard reference\nd=1.0, g=2.0, pb=0.1\nnholes=4, nparticles=4')

plot_expects(ss_flow_4)
plt.title('Optimized reference ensemble\nd=1.0, g=2.0, pb=0.1\nnholes=4, nparticles=4')

plot_expects(ss_flow_5)
plt.title('Optimized reference ensemble\nd=1.0, g=0.5, pb=0.1\nnholes=4, nparticles=4')
```
